## Norm of a vector
The norm of a vector is a measure of its length. There are many ways of defining the length of a vector depending on the metric used (i.e., the distance formula chosen). The most common is called the $L_2$ norm, which is computed according to the distance formula you are probably familiar with from grade school. The $L_2$ norm of a vector $v$ is denoted by $∥v∥_2$ and 

$$\Vert v \Vert_{2} = \sqrt{\sum_i v_i^2}$$

This is sometimes also called Euclidian length and refers to the “physical” length of a vector in one-, two-, or three-dimensional space. The $L_1$ norm, or “Manhattan Distance,” is computed as

$$\Vert v \Vert_{1} = \sum_i |v_i|$$

and is named after the grid-like road structure in New York City. In general, the p-norm, $L_p$, of a vector is

$$\Vert v \Vert_{p} = \sqrt[p]{(\sum_i v_i^p)}$$

The $L_\infty$ norm is the p-norm, where $p=\infty$. The $L_\infty$ norm is written as $||v||_\infty$ and it is equal to the maximum absolute value in $v$.

## Python code for calculating norms

All the linear algebraic operations in python are defined in `numpy.linalg` library. So we have to import this library to implement vector/ matrix operations in python.

```python
import numpy.linalg as la
new_vector = vector_row.T # .T stands for transpose
print("The Transposed vector\n",new_vector)
norm_1 = la.norm(new_vector, 1)
norm_2 = la.norm(new_vector, 2)
norm_inf = la.norm(new_vector, np.inf)
print('L_1 norm is: %.1f'%norm_1)
print('L_2 norm is: %.1f'%norm_2)
print('L_inf norm  is: %.1f'%norm_inf)
```

### `Dot` product
> The dot product of two vectors is the sum of the product of the respective elements in each vector and is denoted by $\cdot$, and $v\cdot w$ is read “v dot w.” This operation can be defined as $v\cdot w = \sum_{i = 1}^{n} v_iw_i$.

### Angle between two vectors
>Angle between two vectors can be calculated from the following relationship
$$v \cdot w = \Vert v \Vert_{2} \Vert w \Vert_{2} \cos{\theta}$$
So the angle between $v$ and $w$ is calculated as

$$\theta=\cos^{-1}\biggl(\dfrac{v\cdot w}{||v||_2\times ||w||_2}\biggr)$$
 An example is given bellow:





```python
v = np.array([[10, 9, 3]])
w = np.array([[2, 5, 12]])
theta = \
    np.arccos(np.dot(v, w.T)/(la.norm(v)*la.norm(w)))
print(theta)
```

### Broadcasting
`Numpy` can handle operations on arrays of different shapes. The smaller array will be extended to match the shape of the bigger one. The advantage is that this is done in C under the hood (like any vectorized operations in `Numpy`). Actually, we used broadcasting in the previous example . The scalar was converted in an array of same shape as $\mathbf{A}$.

Here is another generic example:
\begin{equation}
\begin{bmatrix} A_{1,1} & A_{1,2} \\ A_{2,1} & A_{2,2} \\ A_{3,1} & A_{3,2} \end{bmatrix}+ \begin{bmatrix} B_{1,1} \\ B_{2,1} \\ B_{3,1} \end{bmatrix}\end{equation}
is equivalent to


\begin{equation}
\begin{bmatrix} A_{1,1} & A_{1,2} \\ A_{2,1} & A_{2,2} \\ A_{3,1} & A_{3,2} \end{bmatrix}+ \begin{bmatrix} B_{1,1} & B_{1,1} \\ B_{2,1} & B_{2,1} \\ B_{3,1} & B_{3,1} \end{bmatrix}= \begin{bmatrix} A_{1,1} + B_{1,1} & A_{1,2} + B_{1,1} \\ A_{2,1} + B_{2,1} & A_{2,2} + B_{2,1} \\ A_{3,1} + B_{3,1} & A_{3,2} + B_{3,1} \end{bmatrix}
\end{equation}
where the (3×1) matrix is converted to the right shape (3×2) by copying the first column. `Numpy` will do that automatically if the shapes can match.

Example:

## Creating augmented matrix using `numpy`

It is common to use the augmented matrix while solving system of linear equations. `concantenate` function from `numpy` library is used for this purpose. Suppose $A=\begin{pmatrix}1&1&0\\0&1&0\\1&0&1\end{pmatrix}$ and $y=\begin{pmatrix}1\\2\\1\end{pmatrix}$. Then  $[A:y]=\begin{pmatrix}1&1&0&1\\0&1&0&2\\1&0&1&1\end{pmatrix}$

```python
A = np.array([[1,1,0],
              [0,1,0],
              [1,0,1]])
print('Rank of A:', la.matrix_rank(A))
y = np.array([[1], [2], [1]])
A_y = np.concatenate((A, y), axis = 1)
print('Augmented matrix:\n', A_y)
print("Rank of Augmented matrix is:",la.matrix_rank(A_y))
```

### Inverse Matrices
The matrix inverse of $A$ is denoted $A^{−1}$. It is the matrix that results in the identity matrix when it is multiplied by $A$:
>**Note:**  We will use the Numpy function `linalg.inv()` to calculate the inverse of $A$.

Find the inverse of the matrix $ \mathbf{A}=\begin{bmatrix} 3 & 0 & 2 \\ 2 & 0 & -2 \\ 0 & 1 & 1 \end{bmatrix}$.

```python
A = np.array([[3, 0, 2], [2, 0, -2], [0, 1, 1]]) # assigning the matrix
A_inv = np.linalg.inv(A) # calculating the inverse and stores in A_inv
print(A_inv)
print(np.dot(A_inv,A)) # verify that the product of A and A_inv gives the identity matrix
```

### Solving a system of linear equations

The inverse matrix can be used to solve the equation $\mathbf{Ax}=\mathbf{b}$ using the formula $\mathbf{x}=\mathbf{A}^{-1}\mathbf{b}$. This formula is valid only if $A$ is invertible.

As an example, solve $\begin{cases} y = 2x \\ y = -x +3 \end{cases}$ using matrix inversion method.

**Step 1:** Formulating Matrix form
Our matrix A of weights is: $\mathbf{A}= \begin{bmatrix} 2 & -1 \\ 1 & 1 \end{bmatrix}$ and $\mathbf{b}= \begin{bmatrix} 0 \\ 3 \end{bmatrix}$. So the matrix form will be:

\begin{equation}
\begin{bmatrix} 2 & -1 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}= \begin{bmatrix} 0 \\ 3 \end{bmatrix}
\end{equation}
**Step 2:** Solution using $\mathbf{x}=\mathbf{A}^{-1}\mathbf{b}$


```python
A = np.array([[2, -1], [1, 1]]) # assigning weights
b = np.array([[0], [3]])        # assigning constant column
A_inv = np.linalg.inv(A)        # calculating inverse of the weight matrix
x=A_inv.dot(b)
#x=np.dot(A_inv,b)
print("The solution is:\n",x)
```

## Fun with SVD on the image

```python
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
plt.figure(figsize=(8,6))
plt.imshow(imgmat, cmap='gray');
```

As promised, one line of command is enough to get the singular value decomposition. $U$ and $V$ are the left-hand side and the right-hand side matrices, respectively. 'sigma' is a vector containing the diagonal entries of the matrix $\Sigma$ The other two lines reconstruct the matrix using the first singular value only. You can already guess the rough shape of the original image.

```python
U, sigma, V = np.linalg.svd(imgmat)

reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
plt.imshow(reconstimg, cmap='gray');
```

Let's see what we get when we use the second, third and fourth singular value as well. 

```python
for i in range(2, 4):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

Now we let $i$ run from 5 to 51, using a step width of 5. For $i=50$, we already get a pretty good image!

```python
for i in range(5, 51, 5):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

But how many singular values do we have after all? The following command gives us the number of entries in sigma. As it is the diagonal matrix, it is stored as a vector and we do not save the zero entries. We now output the number of singular values (the length of the vector sigma, containing the diagonal entries), as well as the size of the matrices $U$ and $V$.

```python
print("We have %d singular values." % sigma.shape)
print("U is of size", U.shape, ".") 
print("V is of size", V.shape, ".")
print("The last, or smallest entry in sigma is", sigma[224])
```

As Python stores the whole singular value decomposition, we do not really save space.
